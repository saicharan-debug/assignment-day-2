```python
a = {'x':0, 'y':0}
n = int(input('enter the number'))
for i in range (n):
    move = input ().split(' ')
    if move[0].lower()=='up':
        a['y'] += int(move[1])
    elif move[0].lower()=='down':
        a['y'] -= int(move[1])
    elif move[0].lower()=='left':
        a['x'] -= int(move[1])
    elif move[0].lower()=='right':
        a['x'] += int(move[1])
print(int(round(a['x']**2 + a['y']**2)**0.5))
            
```
